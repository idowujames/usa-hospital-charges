### A personal Project to merge two excel files to only have a list of members who have paid for their dues in 2023

```python
import pandas as pd
```

```python
def wrangle(path):
  cols_to_drop = ['Unnamed: 0','2020/21','2021/2022','Unnamed: 9']
  df = (
            pd.read_csv(path,dtype={'TELEPHONE': str})
            .pipe(lambda df_: df_.drop(columns=[cols for cols in cols_to_drop if cols in df_.columns],axis=1))
            .dropna(how='all')
            .rename(columns=lambda x: x.strip())
            .apply(lambda x: x.str.strip() if x.dtype == "object" else x)
        )
  return df
```

```python
members = wrangle('/content/ANWBN COALITION MEMBERS CIPE (22) - ANWBN MEMBERSHIP PROFILE (1).csv')
print(members.info())
members.head()
```

```python
finan = wrangle('/content/Financial ANWBN MEMBERSHIP 2023 - ANWBN MEMBERSHIP PROFILE.csv')
print(finan.info())
finan.head()
```

```python
right_df = pd.merge(members,finan, on='ASSOCIATION', how='right')
left_df = pd.merge(members,finan, on='ASSOCIATION', how='left')
df = pd.merge(members,finan, on='ASSOCIATION')

print(right_df.shape,left_df.shape,df.shape)
```

```python
df2 = (
    df[df['2022/2023'].notna()]
    .drop(columns=['PRESIDENT_y','ALTERNATE_y','TELEPHONE_y','EMAIL_y'], axis=1)
    .rename(columns=lambda x: x.replace('_x', ''))
)
df2.info()
df2.head()
```

```python
df2.to_excel('Financial Members.xlsx', index=False)
```
